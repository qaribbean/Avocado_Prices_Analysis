#### Part 2. Further Analysis

#### Prices by Type

Investigating the two types of avocado for a difference in price, we can plot a time series of average price:

```python
avocado_date_t=avocado.pivot_table(index='Date', columns='type', aggfunc='mean')['AveragePrice'] #Place the data into a pivot table.
```

```python
avocado_date_t.plot(figsize=(15,5)) #Plot both avocado types by price
plt.ylabel('Average Price')
```

While the price of the Organic is consistently higher than Conventional, one can observe that from February 2017 the price of Conventional rose sharply to almost match Organic in March 2017 (see chart below):

```python
avocado_date_t.plot(figsize=(15,5))
plt.ylabel('Average Price')
plt.text(x='2017-1-15', y=2.04, s='March 2017', color='red', fontsize=12)
plt.vlines(x='2017-3-1', ymin=0.5, ymax=2, color='red', linestyles=':', linewidth=3, label='March 2017')
```

Noticing the sharp rise in February to March 2017, we then also notice that we see the sharpest decrease in price for CONVENTIONAL types every February.  This February price movement is not quite so prevalent in Organic types.

#### Seasonality

Returning to seasonality, based on our first average price chart, we can break out the seasonality further and explore if this trend has existed over the individual years available:

```python
x=[]
for i in range(len(avocado)):
    m=avocado['Date'].loc[i].strftime("%B")
    x.append(m)
```

```python
avocado['month']=x
```

```python
avocado_year=avocado.pivot_table(index='month', columns='year', aggfunc='mean')['AveragePrice']
```

```python
month_order={
    ',':0,
    'January':1,
    'February':2,
    'March':3,
    'April':4,
    'May':5,
    'June':6,
    'July':7,
    'August':8,
    'September':9,
    'October':10,
    'November':11,
    'December':12,
    ',':13
}
```

```python
fig=avocado_year.loc[month_order].plot(figsize=(15,4), xticks=range(0,14), cmap='jet')
plt.text(x=1.1, y=1.9, s='Winter', color='#98F5FF', fontsize=15)
plt.vlines(x=1, ymin=0.8, ymax=2, color='lightblue', linestyles=':', linewidth=3)
plt.text(x=3.1, y=1.9, s='Spring', color='#7FFF00', fontsize=15)
plt.vlines(x=3.0, ymin=0.8, ymax=2, color='#7FFF00', linestyles=':', linewidth=3)
plt.text(x=6.1, y=1.9, s='Summer', color='red', fontsize=15)
plt.vlines(x=6.0, ymin=0.8, ymax=2, color='red', linestyles=':', linewidth=3)
plt.text(x=9.1, y=1.9, s='Autumn', color='orange', fontsize=15)
plt.vlines(x=9.0, ymin=0.8, ymax=2, color='orange', linestyles=':', linewidth=3)
plt.ylabel('Average Price')
```

We can see that in the first full year of data (2015) the Summer-Autumn price increase was essentially non-existent.  In 2016 the upward trend becomes more prominent and in 2017 it is clearly in place.<br> <br>This does now open the question regarding 2018 data being incomplete as to whether average prices continued higher and if the seasonal trend repeated itself.
